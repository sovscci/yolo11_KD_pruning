*Below are the definitions of different the pruning methods *

```python
import torch
import torch.nn.utils.prune as prune
from ultralytics import YOLO

def prune_yolov5_unstructured(model, amount=0.1):
    for module in model.modules():
        # Prune only the inner nn.Conv2d inside YOLOv5's Conv blocks
        if isinstance(module, torch.nn.Conv2d):
            prune.l1_unstructured(module.conv, name='weight', amount=amount)
            prune.remove(module.conv, 'weight')  # Make pruning permanent
    return model

def prune_model(model, amount=0.1): 
    print(f"prune_model : {amount}")
    for module in model.modules():
        if isinstance(module, torch.nn.Conv2d):
            prune.l1_unstructured(module, name='weight', amount=amount)
            prune.remove(module, 'weight') 
    return model 

def prune_model_l2(model, amount=0.05):
    for module in model.modules():
        if isinstance(module, torch.nn.Conv2d):
            prune.ln_structured(module, name='weight', amount=amount, n=2, dim=0)
            prune.remove(module, 'weight') 
    return model
def prune_model_filters(model, amount=0.2):
    for module in model.modules():
        if isinstance(module, torch.nn.Conv2d):
            prune.ln_structured(module, name='weight', amount=amount, n=1, dim=0)  # Removes filters
            prune.remove(module, 'weight')
    return model
def prune_model_channels(model, amount=0.2):
    for module in model.modules():
        if isinstance(module, torch.nn.Conv2d):
            prune.ln_structured(module, name='weight', amount=amount, n=1, dim=1)  # Removes channels
            prune.remove(module, 'weight')
    return model
def global_prune(model, amount=0.05):
    parameters_to_prune = []
    for module in model.modules():
        if isinstance(module, torch.nn.Conv2d) or isinstance(module, torch.nn.Linear):
            parameters_to_prune.append((module, 'weight'))

    prune.global_unstructured(parameters_to_prune, pruning_method=prune.L1Unstructured, amount=amount)

    for module, _ in parameters_to_prune:
        prune.remove(module, 'weight')

    return model
def iterative_prune(model, initial_amount=0.1, iterations=5):
    for i in range(iterations):
        amount = initial_amount * (i + 1) / iterations  # Gradual increase
        model = prune_model(model, amount)
    return model
def random_prune(model, amount=0.05):
    for module in model.modules():
        if isinstance(module, torch.nn.Conv2d):
            prune.random_unstructured(module, name='weight', amount=amount)
            prune.remove(module, 'weight')
    return model

def structured_prune_model(model, amount=0.2, dim=1):
    """
    Apply structured pruning to all Conv2d layers of the model.
    
    Args:
        model: The model to prune.
        amount: Fraction of channels/filters to prune.
        dim: Dimension to prune along:
            - dim=0: Prunes output channels (filters).
            - dim=1: Prunes input channels.
    """
    for module in model.modules():
        if isinstance(module, torch.nn.Conv2d):
            prune.ln_structured(module, name='weight', amount=amount, n=2, dim=dim)
            prune.remove(module, 'weight')  # Remove reparameterization, make pruning permanent
    return model

```
